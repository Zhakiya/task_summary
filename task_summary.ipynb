```python
# Contoh Analisis Data Kinerja Mahasiswa per Mata Kuliah
# Install library jika belum tersedia (biasanya tidak perlu di Colab)
# !pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# 1. Membuat Dataset Contoh (Simulasi Data)
data = {
    'NIM': ['001', '002', '003', '004', '005'],
    'Nama': ['Anna', 'Budi', 'Citra', 'Doni', 'Eka'],
    'Matematika': [75, 60, 88, 45, 90],
    'Biologi': [82, 70, 65, 78, 92],
    'Fisika': [60, 55, 70, 82, 75]
}

df = pd.DataFrame(data)
print("Data Mentah:")
display(df)

# 2. Analisis per Mata Kuliah (Kategori)
def analyze_subject(df, subject):
    avg = df[subject].mean()
    min_score = df[subject].min()
    max_score = df[subject].max()
    under_70 = df[df[subject] < 70].shape[0]

    return {
        'Mata Kuliah': subject,
        'Rata-Rata': round(avg, 2),
        'Nilai Terendah': min_score,
        'Nilai Tertinggi': max_score,
        'Siswa <70': under_70
    }

# Hasil Analisis per Kategori
results = []
for subject in ['Matematika', 'Biologi', 'Fisika']:
    results.append(analyze_subject(df, subject))

result_df = pd.DataFrame(results)
print("\nAnalisis per Mata Kuliah:")
display(result_df)

# 3. Visualisasi
plt.figure(figsize=(10, 5))
result_df.plot(x='Mata Kuliah', y=['Rata-Rata', 'Nilai Terendah', 'Nilai Tertinggi'],
               kind='bar', title='Perbandingan Nilai per Mata Kuliah')
plt.ylabel('Nilai')
plt.show()
```
